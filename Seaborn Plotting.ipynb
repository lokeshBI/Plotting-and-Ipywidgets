## Seaborn Plotting
### Statistical Analysis plotting
#### scatterplot() and lineplot() using seaborn's default relplot()

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head(10)
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, kind = 'scatter')
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, kind = 'scatter',hue= 'smoker')
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, kind = 'scatter',hue= 'smoker', style = 'smoker')
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, kind = 'scatter',hue= 'smoker', style = 'size')
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, kind = 'scatter',hue= 'smoker', style = 'size', size = 'time')
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, kind = 'scatter',hue= 'size')
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, kind = 'scatter',hue = 'size',size = 'size')
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips, kind = 'scatter',hue = 'size',size = 'size', sizes= (15,200))
sns.set(style = 'darkgrid')
```

### lineplot()

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
print(fmri.event.unique())
print(fmri.region.unique())
```

```python
sns.relplot(x = 'timepoint', y = 'signal', data = fmri, kind = 'line')
sns.set(style = 'darkgrid')
```

```python
## you can put "confidence interval = none"
sns.relplot(x = 'timepoint', y = 'signal', data = fmri, kind = 'line', ci = None)
sns.set(style = 'darkgrid')
```

```python
## to turn off aggregations set "estimator = None"
sns.relplot(x = 'timepoint', y = 'signal', data = fmri, kind = 'line', ci = None, estimator = None)
sns.set(style = 'darkgrid')
```

### plotting subsets of data

```python
sns.relplot(x = 'timepoint', y = 'signal', data = fmri, kind = 'line', hue= 'event')
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', data = fmri, kind = 'line', hue= 'region', style= 'event', ci = None)
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', data = fmri, kind = 'line', hue= 'event', style= 'event', ci = None, 
            markers= True, dashes= False)
sns.set(style = 'darkgrid')
```

### plotting with date

```python
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
df.head()
```

```python
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df).fig.autofmt_xdate()
```

### showing multiple reltions with facets
#### col -- columnwise faceting, row -- rowwise faceting

```python
tips.head()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips ,hue = 'smoker',col = 'time', kind = 'scatter')
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips ,hue = 'smoker',col = 'sex',row = 'time', kind = 'scatter')
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x="timepoint", y="signal", hue="subject",
            col="region", row="event", height=3,
            kind="line", estimator=None, data=fmri)
sns.set(style = 'darkgrid')
```

## Plotting with categorical data
### Categorical scatterplots:

###### stripplot() (with kind="strip"; the default) 
###### swarmplot() (with kind="swarm")

### Categorical distribution plots:

###### boxplot() (with kind="box")
###### violinplot() (with kind="violin")
###### boxenplot() (with kind="boxen")

### Categorical estimate plots:

###### pointplot() (with kind="point")
###### barplot() (with kind="bar")
###### countplot() (with kind="count")

```python
tips.head()
```

```python
sns.catplot(x= 'day', y = 'total_bill', data = tips, kind = 'strip')
sns.set(style='darkgrid')
```

```python
sns.catplot(x= 'day', y = 'total_bill', data = tips, kind = 'swarm')
sns.set(style='darkgrid')
```

```python
sns.catplot(x= 'day', y = 'total_bill', data = tips, kind = 'swarm', hue = 'smoker')
sns.set(style='darkgrid')
```

```python
sns.catplot(x= 'smoker', y = 'tip', data = tips, kind = 'swarm')
sns.set(style='darkgrid')
```

```python
## we can change the order as well for smokers 
sns.catplot(x= 'smoker', y = 'tip',order= ['No','Yes'], data = tips, kind = 'swarm')
sns.set(style='darkgrid')
```

```python
sns.catplot(x= 'day', y = 'total_bill', data = tips, kind = 'box')
sns.set(style='darkgrid')
```

```python
sns.catplot(x= 'day', y = 'total_bill', data = tips, kind = 'violin')
sns.set(style='darkgrid')
```

```python
sns.catplot(x= 'day', y = 'total_bill', data = tips, kind = 'violin', hue = 'smoker')
sns.set(style='darkgrid')
```

```python
## rather than separating smokers we can split the plot as well.
sns.catplot(x= 'day', y = 'total_bill', data = tips, kind = 'violin', hue = 'smoker', split = True)
sns.set(style='darkgrid')
```

```python
sns.catplot(x= 'day', y = 'total_bill', data = tips, kind = 'bar', hue = 'smoker')
sns.set(style='darkgrid')
```

```python
sns.catplot(y = 'day', x = 'total_bill', data = tips, kind = 'bar', hue = 'smoker')
sns.set(style='darkgrid')
```

```python
sns.catplot(x= 'day', data = tips, kind = 'count', hue = 'smoker')
sns.set(style='darkgrid')
```

### Distribution plotting

```python
x = np.random.normal(size=100)
print(x)
```

```python
sns.distplot(x)
sns.set(style = 'darkgrid')
```

```python
sns.distplot(x, bins= 20)
sns.set(style = 'darkgrid')
```

```python
sns.distplot(tips.tip, kde = False)  # kde = false will remove the line
sns.set(style = 'darkgrid')
```
